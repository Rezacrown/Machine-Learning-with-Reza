# **Linar Regression**

### **Definisi**

Linear Regression adalah model yang mencari hubungan linier antara variabel input (X) dan output (y) dengan meminimalkan error (selisih antara nilai prediksi dan aktual).

### **Cara Kerja**

Model menghitung garis lurus terbaik yang memprediksi nilai (y) berdasarkan kombinasi linier dari fitur (X). Menggunakan metode **Ordinary Least Squares (OLS)** untuk meminimalkan jumlah kuadrat error.

**Linear Regression Formula**:

$\hat{y} = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \dots + \beta_n x_n$

```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings

warnings.filterwarnings("ignore")
```

```python
# import Dataset
from sklearn.datasets import make_regression 

X, y = make_regression(n_samples=500, n_features=5, noise=0, bias=0, random_state=42)

X.shape, y.shape
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
# split data test and train
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Standarisation Data (X) train and test
SK = StandardScaler()

X_train = SK.fit_transform(X_train, y_train)

X_test = SK.transform(X_test)

X_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# init instance of Model
LR = LinearRegression()

# train Data features (X) with Linear Regression Model
LR.fit(X_train, y_train)

print(LR.get_params())
```

```python
# predict and evalutaion
from sklearn.metrics import mean_squared_error, mean_absolute_error

y_pred = LR.predict(X_test)

# evaluate
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"Mean Absolute Error: {mae}")
```

```python
# ploting data
sns.set()

plt.plot(range(len(y_test)), y_test, label="Actual", color="Blue", linewidth=0.5)
plt.plot(range(len(y_pred)), y_pred, label="Predicted", color="Red" , linewidth=0.5)
plt.legend()

plt.show()
```
